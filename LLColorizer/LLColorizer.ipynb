```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_image = cv2.imread('test_image.jpeg')
```

```python
# Displays the shape of the image  where height = 1080, width = 1920, channel = 3(BGR)
test_image.shape 
```

```python
plt.imshow(test_image)
```

# Note: matplotlib displays an image in RGB while CV displays an image in GBR

```python
# Here, I converted from GBR TO RGB, and increased the figure size
test_image = cv2.cvtColor(test_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize = (15,15))
plt.grid(color='w', linestyle='-', linewidth=2)
plt.imshow(test_image)
```

```python
# Check the value of a pixel of an image at a certian position, we do:
position = (250, 250)
pixel_value = test_image[position]
print(f'The pixel value at position {position} is {pixel_value}')
```

# We get three values, the red, Green, and the Blue  values of the pixel at that position.

```python
# Here we would convert to the gray color space
test_image = cv2.cvtColor(test_image,cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (15,15))
plt.grid(color='w', linestyle='-', linewidth=2)
plt.imshow(test_image)
```

```python
# Resizing the image size.

#height = test_image.shape[0]
#width = test_image.shape[1]
height = 1080
width = 1500
dim = (width, height)
test_image = cv2.resize(test_image, dim, interpolation = cv2.INTER_AREA)
```

```python
plt.figure(figsize = (15,15))
plt.grid(color='w', linestyle='-', linewidth=2)
plt.imshow(test_image)
```

```python
# Check the value of a pixel of an image at a certian position, we do:
position = (250, 250)
pixel_value = test_image[position]
print(f'The pixel value at position {position} is {pixel_value}')
```

# We get a single value because it is in the gray color space.

```python
#This is the highest value of pixel in the gray space.
test_image.max() 
```

```python
#This is the lowest value of pixel in the gray space.
test_image.min() 
```

```python
# Converting rom GRAY color space back to RGB color space we get the following
test_image = cv2.cvtColor(test_image,cv2.COLOR_GRAY2RGB)
plt.figure(figsize = (15,15))
plt.grid(color='w', linestyle='-', linewidth=2)
plt.imshow(test_image)
```

```python
# Shape for the image in the RGB color space.
test_image.shape
```

```python
# Checking the pixel value of the image at a certian position:
position = (250, 250)
pixel_value = test_image[position]
print(f'The pixel value at position {position} is {pixel_value * 20}')
```

# Changing the color of the image, by manipulating the pixel values.

```python
# knowing the maximum and minimum value of a pixel helps us in choosing the range of numbers we should add or subtract to each pixel.
test_image.max()
```

```python
test_image.min()
```

```python
hold = np.zeros_like(test_image)
for i in range(height):
    for j in range(width):
        position = (i, j)
        pixel_value = test_image[position]
        hold[position] = [pixel_value[0] * 0.539 , pixel_value[1] * 0.638, pixel_value[2] * 0.837]
```

### OR

```python
hold = np.zeros_like(test_image)
for i in range(height):
    for j in range(width):
        position = (i, j)
        pixel_value = test_image[position]
        
        if i <= 350 and j <= 350:
            hold[position] = [pixel_value[0] * 0.539 , pixel_value[1] * 0.638, pixel_value[2] * 0.837]
                    
        else:
            hold[position] = [pixel_value[0] + 14 , pixel_value[1] - 2, pixel_value[2] + 9]
```

```python
plt.figure(figsize = (15,15))
plt.grid(color='w', linestyle='-', linewidth=2)
plt.imshow(hold)
```
