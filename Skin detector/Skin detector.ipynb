```python
# Required modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Using hsv
image = cv2.imread("test.JPG")
plt.imshow(image)
```

```python
# converting from BGR TO RGB
#imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#plt.imshow(imageRGB)
# note: do not convert from rgb to hsv, it produces a black image when trying to detect skin
```

```python
# converting RGB to HSV
imageHSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(imageHSV)
```

```python
min_HSV = np.array([0, 58, 30], dtype = "uint8")
max_HSV = np.array([33, 255, 255], dtype = "uint8")

skinRegionHSV = cv2.inRange(imageHSV, min_HSV, max_HSV)
skinHSV = cv2.bitwise_and(image, image, mask = skinRegionHSV) # used in merging images together or portions of it.
plt.figure(figsize=(12,12))
plt.imshow(skinHSV)

```

# using the ycrcb

```python
# Required modules
import cv2
import numpy as np
import matplotlib.pyplot as plt

min_YCrCb = np.array([0,133,77],np.uint8)
max_YCrCb = np.array([235,173,127],np.uint8)


```

```python
image = cv2.imread("test.JPG")
plt.imshow(image)
```

```python
imageYCrCb = cv2.cvtColor(image,cv2.COLOR_BGR2YCR_CB)
plt.imshow(imageYCrCb)
```

```python
skinRegionYCrCb = cv2.inRange(imageYCrCb,min_YCrCb,max_YCrCb)
hold = cv2.bitwise_and(image, image, mask = skinRegionYCrCb)
plt.imshow(hold)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(hold)
```

# Color space histogram

cv2.calcHist(images, channels, mask, histSize, ranges)
images : surce image of type uint8 or float32.
channels : Index of channel for which we calculate histogram.
mask : mask image. To find histogram of full image, it is given as “None”.
histSize : this represents our BIN count. For full scale, we pass [256].
ranges : this is our RANGE. Normally, it is [0,256].


```python
# HSV Color-Space Histogram
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([skinHSV],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    plt.ylim([0,500])
plt.show()
```
